```python
%sql
SELECT * FROM review_sample
```

```python
%sql
SELECT
  text,
  ai_analyze_sentiment(text) AS sentiment
FROM
  review_sample

```

```python
# Alternative way of writing the above using spark.sql
display(
  spark.sql(""" SELECT
    text,
    ai_analyze_sentiment(text)    AS sentiment
  FROM
    review_sample """)  ) 
```

```python
%sql
--Classifying reviews

SELECT 
  text,
  ai_classify(text, array('neutral_comment', 'positive_comment', 'negative_comment' )) AS category
FROM
  review_sample
```

```python
%sql
--Fix Grammar errors

SELECT 
  text,
  ai_fix_grammar(text ) AS fix_grammar_error
FROM
  review_sample
```

```python
%sql
--Extracting information from reviews
SELECT
  text,
  ai_extract(text, array('location', 'timing', 'email', 'phone', 'person', 'condition' )) AS product_info
FROM
  review_sample
```

```python
%sql

-- Handling PII
SELECT
  text,
  ai_mask(text, array('person', 'email', 'location', 'phone')) AS masked_info
  FROM
    review_sample
```

```python
%sql
SELECT
  text,
  ai_summarize(text , 15) AS review_summary
FROM
  review_sample


```

```python
%sql 
SELECT 
  text,
  ai_translate(text, 'es') AS text_esp
FROM  
  review_sample
```

```python
df_all = df.selectExpr(
    "text", "ai_analyze_sentiment(text) as sentiment",
    "ai_extract(text, array('location', 'timing', 'email', 'phone', 'person', 'condition' )) AS product_info",
    "ai_mask(text, array('person', 'email', 'location', 'phone')) AS masked_info",
    "ai_fix_grammar(text ) AS fix_grammar_error",
    "ai_classify(text, array('neutral_comment', 'positive_comment', 'negative_comment' )) AS category",
    "ai_translate(text, 'es') AS text_esp",
    "ai_summarize(text , 15) AS review_summary"   
    
    )
display(df_all)
```
